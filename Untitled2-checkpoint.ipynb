# Analyze the data using matplotlib between state name and active cases

```python
state_active_data=df[['state_name', 'active']]
state_active_data=state_active_data.sort_values(by='active', ascending=False)
```

```python
# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(state_active_data['state_name'], state_active_data['active'], color='royalblue')
plt.xlabel('State')
plt.ylabel('Active Cases')
plt.title('Active COVID-19 Cases by State')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Ensures labels fit within the figure area

# Show the plot
plt.show()
```

# Analyze the data between state, positive , cured, and death

```python
state_data=df[['state_name', 'positive', 'cured', 'death']]
state_data=state_data.sort_values(by='positive', ascending=False)
```

```python
# Create a bar chart
plt.figure(figsize=(12, 6))
bar_width = 0.2
index = range(len(state_data))

plt.bar(index, state_data['positive'], width=bar_width, label='Positive Cases', color='royalblue', align='center')
plt.bar([i + bar_width for i in index], state_data['cured'], width=bar_width, label='Cured Cases', color='limegreen', align='center')
plt.bar([i + 2 * bar_width for i in index], state_data['death'], width=bar_width, label='Death Cases', color='tomato', align='center')

plt.xlabel('State')
plt.ylabel('Number of Cases')
plt.title('COVID-19 Cases by State')
plt.xticks(index, state_data['state_name'], rotation=90)  # Rotate x-axis labels for better readability
plt.legend()
plt.tight_layout()  # Ensures labels fit within the figure area

# Show the plot
plt.show()
```
